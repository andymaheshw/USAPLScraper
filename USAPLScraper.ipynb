```python
from BeautifulSoup import BeautifulSoup,SoupStrainer
import urllib
import pandas as pd
```

```python
r = urllib.urlopen('http://usapl.liftingdatabase.com/competitions').read()
soup = BeautifulSoup(r)
print type(soup)
ls = []
```

```python
for link in BeautifulSoup(r, parseOnlyThese=SoupStrainer('a')):
    if "competitions-view" in str(link):
        print link["href"]
        ls.append(link["href"])
```

```python
for link in ls:
    r = urllib.urlopen('http://usapl.liftingdatabase.com/'+link).read()
    soup = BeautifulSoup(r)
    #print soup.prettify()
    data = []
    table = soup.find('table', attrs={'id':'competition_view_results'})
    table_body = table.find('tbody')
    rows = table_body.findAll('tr')
    print rows
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
    print data
    break
    
```
